# Similarity Search

```python
from qdrant_client import QdrantClient
import os
from dotenv import load_dotenv
load_dotenv(".env")

qclient = QdrantClient(
    url=os.getenv("QDRANT_DB_URL"),
    api_key=os.getenv("QDRANT_API_KEY")
)
```

```python
# collection name
collection_name =  "bpld_dataset"
```

```python
# get images with vectors (a small sample)

qclient.scroll(
    collection_name=collection_name,
    with_vectors=True,
    limit=12
)
```

```python
qclient.retrieve(
    collection_name=collection_name,
    ids=[0],
)[0]
```

```python
# get similar records

qclient.recommend(
    collection_name=collection_name,
    # query="Healthy 220",
    positive=[0],
    with_payload=True,
    limit=3
)
```

```python
qclient.recommend(
    collection_name=collection_name,
    # query="Healthy 220",
    positive=[560],
    with_payload=True,
    limit=3
)
```
